```python
from config import api_key
import requests
import json
import pandas as pd
```

```python
best_picture = pd.read_csv("Resources/best_picture.csv")
best_picture
```

```python
omdb_data = []
def omdb_api(dataframe):
    for index, row in dataframe.iterrows():
        # we need the film + year for the api call
        film = dataframe.loc[index, "film"]
        film = film.replace(" ", "+")
        year = dataframe.loc[index, "year_film"]
        try:
            # define target url
            target_url = f"http://www.omdbapi.com/?apikey={api_key}&t={film}&y={year}"
            # make api call
            response = requests.get(target_url).json()
            omdb_data.append(response)
            print("FILM FOUND")
        except:
            print(f"{film} NOT FOUND")
```

```python
omdb_api(best_picture)
```

```python
omdb_data_df = pd.DataFrame(omdb_data)
```

```python
omdb_data_df.head()
```

```python
omdb_data_df = omdb_data_df.loc[omdb_data_df["Response"]=="True"]
omdb_data_df.head()
```

```python
omdb_data_df.to_csv("Resources/omdb_data.csv", index=False)
```
