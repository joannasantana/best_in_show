```python
import pandas as pd
import numpy as np
```

```python
best_picture = pd.read_csv("Resources/best_picture.csv")
best_picture
```

```python
best_picture["Title/Year"] = best_picture['film'].astype(str) +"-"+ best_picture["year_film"].astype(str)
```

```python
best_picture.head()
```

```python
omdb_data_df = pd.read_csv("Resources/omdb_data.csv")
omdb_data_df
```

```python
omdb_data_df["Title/Year"] = omdb_data_df['Title'].astype(str) +"-"+ omdb_data_df["Year"]
```

```python
omdb_data_df
```

```python
final_df = pd.merge(best_picture, omdb_data_df, on="Title/Year", how="inner")
final_df
```

```python
final_df.columns
```

```python
final_df.drop(["ceremony","Type","Title","Title/Year", "Response", "Error", "Year", "Plot", "Poster", "Website", "totalSeasons", "DVD", "imdbID", "Ratings"], axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df.columns
```

```python
final_df.dtypes
```

```python
final_df["Runtime"].unique()
```

```python
# Remove " min" from Runtime column so that we can change it to an integer
final_df["Runtime"]=final_df["Runtime"].str.replace(" min", "")
final_df["Runtime"]
```

```python
final_df["imdbVotes"].unique()
```

```python
# Remove "," from imdbVotes column so that we can change it to an integer
final_df["imdbVotes"] = final_df["imdbVotes"].str.replace(",","")
final_df["imdbVotes"]
```

```python
final_df["BoxOffice"].unique()
```

```python
# Remove "," and "$" from BoxOffice column so that we can change it to an integer
final_df["BoxOffice"] = final_df["BoxOffice"].str.replace(",","")
final_df["BoxOffice"] = final_df["BoxOffice"].str.replace("$","")
final_df["BoxOffice"]
```

```python
final_df_copy = final_df.copy()
```

```python
final_df_copy["BoxOffice"] = final_df_copy["BoxOffice"].replace(np.nan, 0)
final_df_copy["imdbVotes"] = final_df_copy["imdbVotes"].replace(np.nan, 0)
final_df_copy["Runtime"] = final_df_copy["Runtime"].replace(np.nan, 0)
final_df_copy["year_film"] = final_df_copy["year_film"].replace(np.nan, 0)
final_df_copy["year_ceremony"] = final_df_copy["year_ceremony"].replace(np.nan, 0)
final_df_copy["Metascore"] = final_df_copy["Metascore"].replace(np.nan, 0)
final_df_copy["imdbRating"] = final_df_copy["imdbRating"].replace(np.nan, 0)
```

```python
final_df_copy.dtypes
```

```python
# Runtime to int64
# imdbVotes to int64
# BoxOffice to int64
final_df_copy = final_df_copy.astype({"Runtime":"int64", "imdbVotes":"int64","BoxOffice":"int64"})
```

```python
final_df_copy.dtypes
```

```python
final_df["Genre"].unique()
```

```python
final_df_copy[["Genre1","Genre2","Genre3"]] = final_df_copy["Genre"].str.split(', ', expand=True)
```

```python
final_df_copy.head()
```

```python
final_df_copy.columns
```

```python
final_df_copy[["Genre1", "Genre2", "Genre3"]]
```

```python
final_df_copy = final_df_copy.drop(["Genre"], axis=1)
final_df_copy.columns
```

```python
final_df_copy.dtypes
```

```python
final_df_copy.to_csv("Resources/final_df.csv", index=False)
```
