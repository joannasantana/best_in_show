This notebook's purpose is to run an API call to gather more data on our Best Picture nominees and winners. We also run API calls on movies anticipated to be nominated at the 2024 Oscars and for fun we've also done API calls on older movies that our group feels were overlooked and did not receive a Best Picture nomination.

```python
from config import api_key
import requests
import json
import pandas as pd
```

```python
# Read in the best_picture csv as a dataframe
best_picture = pd.read_csv("Resources/best_picture.csv")
best_picture
```

```python
# Create a function that will do the OMDB API call
omdb_data = []
def omdb_api(dataframe):
    for index, row in dataframe.iterrows():
        # we need the film + year for the api call
        film = dataframe.loc[index, "film"]
        film = film.replace(" ", "+")
        year = dataframe.loc[index, "year_film"]
        try:
            # define target url
            target_url = f"http://www.omdbapi.com/?apikey={api_key}&t={film}&y={year}"
            # make api call
            response = requests.get(target_url).json()
            omdb_data.append(response)
            print("FILM FOUND")
        except:
            print(f"{film} NOT FOUND")
```

# Run API call on historical Best Picture dataset

```python
# Run the API call on all movies in the best_picture dataframe 
omdb_api(best_picture)
```

```python
# Save the API call results to a dataframe
omdb_data_df = pd.DataFrame(omdb_data)
```

```python
# Preview the dataframe
omdb_data_df.head()
```

```python
# Remove any rows where Response = False, since these movies had no data to be found
omdb_data_df = omdb_data_df.loc[omdb_data_df["Response"]=="True"]
omdb_data_df.head()
```

```python
# Use the shape attribute to see how many movies had API call results
omdb_data_df.shape
```

```python
# Save the API call results to a csv
omdb_data_df.to_csv("Resources/omdb_data.csv", index=False)
```
