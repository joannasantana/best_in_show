This notebook is for merging and cleaning data. We need to merge our historic Oscars with the results of our API call.

```python
import pandas as pd
import numpy as np
```

# Clean and merge best_picture dataframe with the omdb_data dataframe

```python
# Load in CSV files
best_picture = pd.read_csv("Resources/best_picture.csv")
best_picture
```

```python
# Load in CSV files
omdb_data = pd.read_csv("Resources/omdb_data.csv")
omdb_data
```

```python
# Create a Title/Year column so that each dataframe has a common column for merging
best_picture["Title/Year"] = best_picture['film'].astype(str) +"-"+ best_picture["year_film"].astype(str)
omdb_data["Title/Year"] = omdb_data['Title'].astype(str) +"-"+ omdb_data["Year"].astype(str)
```

```python
best_picture.head()
```

```python
omdb_data.head()
```

```python
# Merge the two datframes
combined_df = pd.merge(best_picture, omdb_data, on="Title/Year", how="inner")
combined_df.head()
```

```python
# Review the columns
combined_df.columns
```

```python
combined_df["Awards"]
```

```python
# Check data types
combined_df.dtypes
```

```python
# We want Runtime to be an integer. Check current values for this column
combined_df["Runtime"].unique()
```

```python
# Remove " min" from Runtime column so that we can change it to an integer
def Runtime_cleaning(dataframe):
    dataframe["Runtime"]=dataframe["Runtime"].str.replace(" min", "")
    print(dataframe["Runtime"].unique())
Runtime_cleaning(combined_df)
```

```python
# We want imdbVotes to be an integer. Check current values for this column
combined_df["imdbVotes"].unique()
```

```python
# Remove "," from imdbVotes column so that we can change it to an integer
def imdbVotes_cleaning(dataframe):
    dataframe["imdbVotes"] = dataframe["imdbVotes"].str.replace(",","")
    print(dataframe["imdbVotes"].unique())
imdbVotes_cleaning(combined_df)
```

```python
# We want BoxOffice to be an integer. Check current values for this column
combined_df["BoxOffice"].unique()
```

```python
# Remove "," and "$" from BoxOffice column so that we can change it to an integer
def BoxOffice_cleaning(dataframe):
    dataframe["BoxOffice"] = dataframe["BoxOffice"].str.replace(",","")
    dataframe["BoxOffice"] = dataframe["BoxOffice"].str.replace("$","")
    print(dataframe["BoxOffice"].unique())
BoxOffice_cleaning(combined_df)
```

```python
# We cannot drop all rows with Nan data, as this will not leave us with much of a dataset
# Replace NaN values instead
combined_df["BoxOffice"] = combined_df["BoxOffice"].replace(np.nan, 0)
combined_df["imdbVotes"] = combined_df["imdbVotes"].replace(np.nan, 0)
combined_df["Runtime"] = combined_df["Runtime"].replace(np.nan, 0)
combined_df["year_film"] = combined_df["year_film"].replace(np.nan, 0)
combined_df["year_ceremony"] = combined_df["year_ceremony"].replace(np.nan, 0)
combined_df["Metascore"] = combined_df["Metascore"].replace(np.nan, 0)
combined_df["imdbRating"] = combined_df["imdbRating"].replace(np.nan, 0)
```

```python
# Reminder of current data types
combined_df.dtypes
```

```python
# Change Runtime, imdbVotes, and BoxOffice to int64.
combined_df = combined_df.astype({"Runtime":"int64", "imdbVotes":"int64","BoxOffice":"int64"})
```

```python
# Check data types
combined_df.dtypes
```

```python
# Values for "Genre" may contain up to three values
combined_df["Genre"].value_counts()
```

```python
# Split Genre column into three columns
def genre_splitting(dataframe):
    dataframe[["Genre1","Genre2","Genre3"]] = dataframe["Genre"].str.split(', ', expand=True)
    dataframe.drop(columns="Genre", inplace=True)
    print(dataframe.columns)
genre_splitting(combined_df)
```

```python
# View the new columns
combined_df[["Genre1", "Genre2", "Genre3"]]
```

# Add in data of Golden Globes best picture winners

```python
globes = pd.read_csv("Resources/golden_globe_awards.csv")
globes.head()
```

```python
# View the categories to see which may equate to the Oscars Best Picture category
globes["category"].unique()
```

```python
# Limit the dataframe to the applicable categories
globes_categories = ['Picture', 'Best Motion Picture - Foreign Language', 'Best Motion Picture - Drama',
'Best Motion Picture - Musical or Comedy', 'Foreign Film - Foreign Language', 'Picture - Musical',
'Picture - Comedy']
globes = globes[globes["category"].isin(globes_categories)]
globes.head()
```

```python
# Create a Title/Year column so that each dataframe has a common column for merging
globes["Title/Year"] = globes['nominee'] +"-"+ globes["year_film"].astype(str)
globes.head()
```

```python
# Drop columns that won't be needed
globes = globes[["Title/Year", "win"]]
globes.head()
```

```python
# Combine globes dataframe 
combined_df_with_globes = pd.merge(combined_df, globes, on="Title/Year", how="inner")
combined_df_with_globes.head()
```

```python
# Drop columns so that the only ones remaining are those that we will potentially use in our model
combined_df_with_globes.drop(["ceremony", "Type", "Title", "Response", "Error", "Year", "Plot", "Poster", "Website",
               "totalSeasons", "DVD", "imdbID", "Ratings"], axis=1, inplace=True)
combined_df_with_globes.head()
```

```python
combined_df_with_globes.columns
```

```python
# Clarify Globes vs Oscar winner column names
combined_df_with_globes = combined_df_with_globes.rename(columns={"winner":"OscarsWinner", "win": "GlobesWinner"})
```

```python
combined_df_with_globes.columns
```

```python
# Since the Golen Globes only have 1944-2020 data, our dataset will be more limited from >500 movies to just 220 movies
combined_df_with_globes.shape
```

```python
# Drop columns so that the only ones remaining are those that we will potentially use in our model
combined_df.drop(["ceremony", "Type", "Title", "Title/Year","Response", "Error", "Year", "Plot", "Poster", "Website",
               "totalSeasons", "DVD", "imdbID", "Ratings"], axis=1, inplace=True)
combined_df.head()
```

```python
combined_df.columns
```

```python
# Rename "winner" to "OscarWinner" column
combined_df = combined_df.rename(columns={"winner":"OscarsWinner"})
```

```python
combined_df.columns
```

```python
# Save dataframes to csv files
combined_df.to_csv("Resources/combined_clean.csv", index=False)
combined_df_with_globes.to_csv("Resources/combined_with_globes_clean.csv", index=False)
```
